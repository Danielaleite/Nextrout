# PlotlyVisualizationModule 1: *Barycenters and Grid*.

## The Grid

```python
folder_name='./PVM_data/1_b14_6dv_sf14_rect_cnstrect_cnst/'
file3_ = open(folder_name+"grid.dat", "r")
grid_ = file3_.readlines()
file3_.close()
n_nodes=int(grid_[0])
#print(n_nodes)
grid_nodes=grid_[2:2+n_nodes]
grid_triangles=grid_[2+n_nodes:]

```

```python

# Cleaning the data, extracting the positions of the vertices of the triangles and storing them in a dict, converting them into floats
dict_triang={}
n=1
pos_triang=[line[:-1].split(' ') for line in grid_nodes]
for line_ in pos_triang:
    dict_triang[str(n)]=[]
    line_ = list( dict.fromkeys(line_) )
    if len(line_)<2:
        line_.append(line_[0])
    for i in line_:
        dict_triang[str(n)].append(float(i))
    n+=1
for key_ in dict_triang.keys():
    dict_triang[key_]=np.array(dict_triang[key_])
```

```python
#Defining the graphs and assigning the positions as attributes
G_triang=nx.Graph()
G_triang.add_nodes_from(dict_triang.keys())
for n, p in dict_triang.items():
    G_triang.node[n]['pos'] = p
    #X.nodes[n]['weight'] = weights[int(n)-1]
pos=nx.get_node_attributes(G_triang,'pos')
```

```python
enum_vertex_triang=list(range(1,len(G_triang.nodes())+1))
```

```python
dict_seq={}
n=1
pos_=[line[:-1].split(' ') for line in grid_triangles]
for line_ in pos_:
    dict_seq[str(n)]=[]
    line_ = list( dict.fromkeys(line_) )
    #print(line_)
    if len(line_)>3:
        line_.pop(3)
    for i in line_:
        dict_seq[str(n)].append(str(i))
    n+=1
for key_ in dict_seq.keys():
    dict_seq[key_]=np.array(dict_seq[key_])
```

```python
for i in dict_seq.keys():
    #print(list(itertools.combinations(list(dict_seq[i]), 2)))
    G_triang.add_edges_from(itertools.combinations(list(dict_seq[i]), 2))
    #fig = go.Figure(go.Scatter(x=[0,1,2,0], y=[0,2,0,0], fill="toself"))
#fig.show()
all_cliques= nx.enumerate_all_cliques(G_triang)
triad_cliques=[x for x in all_cliques if len(x)==3 ]
triad_cliques
```

```python
node_trace_triang = go.Scatter(
    x=[],
    y=[],
    text=[],#enum_verteG_triang,
    mode='markers',
    #hoverinfo='text',
    marker=dict(
        color='red',
        #showscale=True,
        # colorscale options
        #'Greys' | 'YlGnBu' | 'Greens' | 'YlOrRd' | 'Bluered' | 'RdBu' |
        #'Reds' | 'Blues' | 'Picnic' | 'Rainbow' | 'Portland' | 'Jet' |
        #'Hot' | 'Blackbody' | 'Earth' | 'Electric' | 'Viridis' |
        colorscale='YlGnBu',
        #reversescale=True,
        #color='skyblue',
        size=10,
        line=dict(width=2)))

for node in G_triang.nodes():
    x, y = G_triang.node[node]['pos']
    node_trace_triang['x'] += tuple([x])
    node_trace_triang['y'] += tuple([y])
    

edge_trace_triang = go.Scatter(
    x=[],
    y=[],
    line=dict(width=0.5,color='#888'),
    hoverinfo='none',
    mode='lines')

for edge in G_triang.edges():
    x0, y0 = G_triang.node[edge[0]]['pos']
    x1, y1 = G_triang.node[edge[1]]['pos']
    edge_trace_triang['x'] += tuple([x0, x1, None])
    edge_trace_triang['y'] += tuple([y0, y1, None])

fig5 = go.Figure(data=[edge_trace_triang, node_trace_triang],
                 layout=go.Layout(paper_bgcolor='rgba(0,0,0,0)',
    plot_bgcolor='rgba(0,0,0,0)',
                title='<br>',
                titlefont=dict(size=16),
                showlegend=True,
                hovermode='closest',
                autosize=False,
                width=900,
                height=800,
                margin=dict(b=20,l=10,r=5,t=40),
                xaxis=dict(showgrid=True, zeroline=True, showticklabels=True),
                yaxis=dict(showgrid=True, zeroline=True, showticklabels=True)))
                 
iplot(fig5, filename='networkx')
```

## Extracting the barycenter positions 

```python

file1_ = open(folder_name+"graph_cell.dat", "r")
graph_coord_triang = file1_.readlines()
file1_.close()
n_nodes=int(graph_coord_triang[0][0:12])
graph_coordinates=graph_coord_triang[2:2+n_nodes]
```

```python
# Cleaning the data, extracting the positions and storing them in a dict, converting them into floats
bar_pos={}
n=1
bar_pos_list=[line[:-1].split(' ') for line in graph_coordinates]
#print(bar_pos_list)
for line_ in bar_pos_list:
    #print(line_)
    bar_pos[str(n)]=[]
    line_ = list( dict.fromkeys(line_) )
    line_.remove('')
    #print(line_)
    if len(line_)<3:
        line_.pop(1)
        line_.append(line_[0])
    else:
        line_.pop(2)
    for i in line_:
        bar_pos[str(n)].append(float(i))
    n+=1
    #print('gone',line_)
#print(bar_pos_list)
#From list to np.array
for key_ in bar_pos.keys():
    bar_pos[key_]=np.array(bar_pos[key_])
```

```python
len(bar_pos)
```

## Extracting the weights 

```python
file2_ = open(folder_name+"opt_tdens.dat", "r")
opt_tdens = file2_.readlines()
file2_.close()
```

```python

# Cleaning the data, extracting the WEIGHTS and storing them in a dict, converting them into floats
dict_weights={}
weights=[]
n=1
weights_=[line[:-1].split(' ') for line in opt_tdens[3:]]
for line_ in weights_:
    line_ = list( dict.fromkeys(line_) )
    line_.remove('')
    if line_[0]!='time':
        num_=float(line_[1])
        dict_weights[str(n)]=num_
        weights.append(num_)
    n+=1
```

```python
#Defining the graphs and assigning the positions as attributes
G_bar=nx.Graph()
G_bar.add_nodes_from(bar_pos.keys())
for n, p in bar_pos.items():
    G_bar.node[n]['pos'] = p
    G_bar.nodes[n]['weight'] = weights[int(n)-1]
pos=nx.get_node_attributes(G_bar,'pos')
```

```python
max_=max(weights)
```
